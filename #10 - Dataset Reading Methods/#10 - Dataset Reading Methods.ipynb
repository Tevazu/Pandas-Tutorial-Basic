```python
import pandas as pd
```

```python
dataset = pd.read_csv("Usvideos.csv") # We read the file
```

```python
dataset
```

#### ○ At the bottom, you can see how many rows and columns are there, let's remove "video_id" and "trending_date" columns and check the numbers again:

```python
newdataset1 = dataset.drop(["video_id", "trending_date"], axis = 1)
```

```python
newdataset1
```

#### ○ As you can see, number of columns are reduced by 2
### Let's create another dataset from the data of this dataset:

```python
newdataset1.to_csv("USvideosNew.csv") #We've created a new file in the same file of this jupyter file as "USvideosNew.csv"
```

### If you open new file, you'll see the two columns will be deleted but there will be index numbers. We don't want it so let's create another one which we deleted them:

```python
newdataset1.to_csv("USvideosNew2.csv", index = False) # "index = False" doesn't show index numbers
```

#### When we write it like "new[0]" it will be a table but there is a problem, the columns "OrderData, Region, Rep, Items, Units, UnitCost. and Total" are become values, not columns. That's because we should give one more parameter in the code we took data from the website. Check the example and continue below to see the correct form.

```python
new[0]
```

### Correct form:
#### We should add an header with value "0". That makes the 0th index values column names.

```python
new = pd.read_html("https://contextures.com/xlsampledata01.html", header = 0)
```

```python
new[0] # Now it's okay
```
