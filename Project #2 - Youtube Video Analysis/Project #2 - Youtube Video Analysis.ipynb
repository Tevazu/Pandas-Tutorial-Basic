##### (Challenge) Sort the videos from the highest to the lowest
##### Please check the explanations given right side of the codes, if you don't understand

```python
def likesDislikesRatio(likes, dislikes):
    likesList = []  # We create a list for likes
    for key, value in list(likes.iteritems()):  # ".iteritems()" takes every index and it's value, then returns them into a tuple 
            likesList.append(value)  # Adding every like value to likeList
    
    dislikesList = [] # We create a list for dislikes
    for key, value in list(dislikes.iteritems()): 
            dislikesList.append(value)  # Adding every dislike value to dislikeList
            
    likedislike = list(zip(likesList, dislikesList))  # Adding two list with "zip()"" function into another list
    ratio = []  #A list for ratios
    for like, dislike in likedislike:
            if (like + dislike) == 0:  # If the result equals "0" then it will return the ratio as "0"
                    ratio.append(0)
            else:    
                    ratio.append(like / (like + dislike))
    return ratio

dataset["likes_dislikes_ratios"] = likesDislikesRatio(dataset["likes"], dataset["dislikes"])
    
```
