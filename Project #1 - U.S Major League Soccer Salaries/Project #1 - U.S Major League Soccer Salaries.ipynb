***
##### Find the mean of all salaries

```python
dataset["base_salary"].mean()
```

***
##### Find the highest salary

```python
dataset["base_salary"].max()
```

***
##### Find the last name of the player with the highest compensation

```python
dataset[dataset["guaranteed_compensation"] == dataset["guaranteed_compensation"].max()]["last_name"].iloc[0]
```

***
##### Find the position the football player named "Gonzalez Pirez" plays.

```python
dataset[dataset["last_name"] == "Gonzalez Pirez"]["position"].iloc[0]
```

***
##### Find average salaries for positions by grouping players by position

```python
dataset.groupby("position").mean()
```

***
##### Calculate how many different positions are in the dataset.

```python
len(dataset.groupby("position"))
```

***
##### Find players with "son" in their last name

```python
def find(last_name):
    if "son" in last_name.lower():
        return True
    return False
```

```python
dataset[dataset["last_name"].apply(find)]
```
